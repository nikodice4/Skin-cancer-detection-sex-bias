```python
# Here are all of our imports used for data cleaning and manipulation

import pandas as pd
import numpy as np
import os
import shutil
```

```python
df_pad_ufes = pd.read_csv("/Users/niko/Documents/uni/6. semester/bsc/Project/BSc-Project/data/df_pad_ufes2.csv")
df_pad_ufes = df_pad_ufes['img_id'].tolist()
df_pad_ufes
```

```python
# allmasks = os.listdir("allMasks")
# allmasks
directory_path = "allMasks"
allmasks = os.listdir(directory_path)

# Print the list of directory contents

allmasks
```

```python
# Extract base filenames without extension from list1
base_filenames_list1 = [filename.split(".")[0] for filename in df_pad_ufes]

# Extract base filenames without extension from list2
base_filenames_list2 = [filename.split("_mask.png")[0] for filename in allmasks]

set1 = set(base_filenames_list1)
set2 = set(base_filenames_list2)

common_elements = set1.intersection(set2)
missing_elements = set1.difference(set2)


# Check if the sets are equal
if common_elements:
    print("There are common elements between the lists:", common_elements)
else:
    print("There are no common elements between the lists.")

print(sorted(common_elements))
print(missing_elements)
```

```python
missing_elements = list(missing_elements)
df = pd.DataFrame(missing_elements, columns=['img_id of missing masks'])

df.sort_values(by='img_id of missing masks', inplace=True)
df.to_csv("missing_masks.csv")
df
```
